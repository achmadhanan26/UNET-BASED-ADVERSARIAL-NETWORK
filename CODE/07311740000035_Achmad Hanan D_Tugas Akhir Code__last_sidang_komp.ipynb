ARSITEKTUR U_NET CODE

```python
config =  tf.compat.v1.ConfigProto()
config.gpu_options.per_process_gpu_memory_fraction = 0.7
set_session(tf.compat.v1.Session(config=config))

K.set_image_data_format('channels_last')  # TF dimension ordering in this code
kinit = 'glorot_normal'

def conv2d_block(n_filters, kernel_size, name):
    model = tf.keras.Sequential()
    for i in range(2):
        model.add(Conv2D(n_filters, kernel_size = (kernel_size,kernel_size), kernel_initializer=kinit, padding="same", name=name+'_conv2d_block{}'.format(i)))
        model.add(BatchNormalization(name=name + '_{}_bn'.format(i)))
        model.add(Activation('relu', name=name + '_{}_act'.format(i)))
    return model

def upsample(outdim, kernel_size):
    model = tf.keras.Sequential()
    model.add(Conv2DTranspose(outdim, kernel_size, strides=2, kernel_initializer=kinit, padding='same'))  
    return model

class U_net(tf.keras.Model):
    def __init__(self):
        super(U_net, self).__init__()

        self.conv_block1 = conv2d_block(64, 3, "conv_block1")
        self.pool1 = MaxPooling2D(pool_size=(2, 2))

        self.conv_block2 = conv2d_block(128, 3, "conv_block2")
        self.pool2 = MaxPooling2D(pool_size=(2, 2))

        self.conv_block3 = conv2d_block(256, 3, "conv_block3")
        self.pool3 = MaxPooling2D(pool_size=(2, 2))

        self.conv_block4 = conv2d_block(512, 3, "conv_block4")
        self.pool4 = MaxPooling2D(pool_size=(2, 2))

        self.bottleneck = conv2d_block(1024, 3, "bottleneck")

        self.up6 = upsample(512, 2)
        self.conv_block5 = conv2d_block(512, 3, "conv_block5")

        self.up7 = upsample(256, 2)
        self.conv_block6 = conv2d_block(256, 3, "conv_block6")

        self.up8 = upsample(128, 2)
        self.conv_block7 = conv2d_block(128, 3, "conv_block7")

        self.up9 = upsample(64, 2)
        self.conv_block8 = conv2d_block(64, 3, "conv_block8")

        self.final_layer = Conv2D(1, (1,1), activation='sigmoid')
      
    def call_model(self):
        global input_size
        ## defining the input shape
        print(input_size)
        inputs = Input(shape=input_size)

        x = inputs
        print('input', x.shape)
        
        c1 = self.conv_block1(x)
        p1 = self.pool1(c1)
        print('c1 {} p1 {}'.format(c1.shape,p1.shape))
        
        c2 = self.conv_block2(p1)
        p2 = self.pool2(c2)
        print('c2 {} p2 {}'.format(c2.shape,p2.shape))

        c3 = self.conv_block3(p2)
        p3 = self.pool3(c3)
        print('c3 {} p3 {}'.format(c3.shape,p3.shape))

        c4 = self.conv_block4(p3)
        p4 = self.pool4(c4)
        print('c4 {} p4 {}'.format(c4.shape,p4.shape))
        
        c5 = self.bottleneck(p4)
        print('c5 {}'.format(c5.shape))
    
        u1 = self.up6(c5)
        up6 = concatenate([u1, c4])
        print('up6 {}'.format(up6.shape))
        c6 = self.conv_block5(up6)
        print('c6', c6.shape)

        u2 = self.up7(c6)
        up7 = concatenate([u2, c3])
        print('up7 {}'.format(up7.shape))
        c7 = self.conv_block6(up7)
        print('c7', c7.shape)

        u3 = self.up8(c7)
        up8 = concatenate([u3, c2])
        print('up8 {}'.format(up8.shape))
        c8 = self.conv_block7(up8)
        print('c8', c8.shape)

        u4 = self.up9(c8)
        up9 = concatenate([u4, c1])
        c9 = self.conv_block8(up9)
        print('c9', c9.shape)
        
        final_layer = self.final_layer(c9)
        print('final layer', final_layer.shape)
        
        return tf.keras.Model(inputs=inputs, outputs=final_layer)
```

ARSITEKTUR JARINGAN DISCRIMINATOR

```python
class resblock(tf.keras.layers.Layer):
    def __init__(self, filter, filter_size):
        super(resblock, self).__init__()
        self.filters = filter
        self.size = filter_size
        self.inp = inp
        self.initializer = tf.random_normal_initializer(0., 0.02)
        self.konvolusi1 = tf.keras.layers.Conv2D(self.filters, 1, strides=1, padding='same', kernel_initializer=self.initializer)
        self.batchNorm0 = tf.keras.layers.BatchNormalization()
        self.konvolusi2 = tf.keras.layers.Conv2D(self.filters, self.size, strides=1, padding='same', kernel_initializer=self.initializer)
        self.batchNorm1 = tf.keras.layers.BatchNormalization()
        self.Relu1 = tf.keras.layers.Activation('relu')
        self.konvolusi3 = tf.keras.layers.Conv2D(self.filters, self.size, strides=1, padding='same', kernel_initializer=self.initializer)
        self.batchNorm2 = tf.keras.layers.BatchNormalization()
        self.batchNorm3 = tf.keras.layers.BatchNormalization()
        self.Relu2 = tf.keras.layers.Activation('relu')
    
    def call(self, inp):
    
        conv1 = self.konvolusi1(inp)
        #batchnorm0 = self.batchNorm1(conv1)
        conv2 = self.konvolusi2(conv1)
        batchnorm1 = self.batchNorm1(conv2)
        relu1 = self.Relu1(batchnorm1)
        conv3 = self.konvolusi3(relu1)
        batchnorm2 = self.batchNorm2(conv3)
        addlayer = tf.keras.layers.Add()([conv1,batchnorm2])
        batchnorm3 = self.batchNorm3(addlayer)
        relu2 = self.Relu2(batchnorm3)

        return relu2

class Discriminator(tf.keras.Model):
    def __init__(self):
        super(Discriminator, self).__init__()

        self.initializer = tf.random_normal_initializer(0., 0.02)
        self.batchnorm1 = tf.keras.layers.BatchNormalization()
        
        self.resblock1 = resblock(8,7)
        self.resblock2 = resblock(8,3)
        self.resblock3 = resblock(16,3)
        self.resblock4 = resblock(32,3)
        self.resblock5 = resblock(64,3)
        self.resblock6 = resblock(64,1)
        self.resblock7 = resblock(64,3)
        self.resblock8 = resblock(64,1)
        self.resblock9 = resblock(64,3)
        
        self.pool1 = tf.keras.layers.AveragePooling2D(pool_size=(2, 2), strides=2, padding='same')
        self.pool2 = tf.keras.layers.AveragePooling2D(pool_size=(2, 2), strides=2, padding='same')
        self.pool3 = tf.keras.layers.AveragePooling2D(pool_size=(2, 2), strides=2, padding='same')
        self.pool4 = tf.keras.layers.AveragePooling2D(pool_size=(2, 2), strides=2, padding='same')
        self.pool5 = tf.keras.layers.AveragePooling2D(pool_size=(2, 2), strides=2, padding='same')
        
        self.last = tf.keras.layers.Flatten()   
        
    def call_model(self):
        global inp
        
        inp_lay = tf.keras.layers.Input(shape=[128,128,1], name='input_image')
        tar = tf.keras.layers.Input(shape=[128,128,1], name='target_image')

        input1 = tf.keras.layers.concatenate([inp_lay, tar], axis=3)
        
        layer1 = self.resblock1(input1)
        pool1 = self.pool1(layer1)
        
        layer2 = self.resblock2(pool1)
        pool2 = self.pool2(layer2)
        
        layer3 = self.resblock3(pool2)
        pool3 = self.pool3(layer3)
        
        layer4 = self.resblock4(pool3)
        pool4 = self.pool4(layer4)

        layer5 = self.resblock5(pool4)
        pool5 = self.pool5(layer5)
        
        #multiple resblock
        layer6 = self.resblock6(pool5)
        layer7 = self.resblock7(layer6)
        layer8 = self.resblock8(layer7)
        layer9 = self.resblock9(layer8)
        
        final_layer = self.last(layer9)
        return tf.keras.Model(inputs=[inp_lay, tar], outputs=final_layer)
```

```python
class GAN_GGO(tf.keras.Model):
    def __init__(self, discriminator, generator, Lambda, Batch_size):
        super(GAN_GGO, self).__init__()
        self.discriminator = discriminator
        self.generator = generator
        self.LAMBDA = Lambda
        self.batch_size = Batch_size
    
    def generate_images(self, model, test_input, tar):
        global prediction

        prediction = model(test_input, training=True)

        display_list = [test_input[0], tar[0], prediction[0]]
        title = ['Input Image', 'Ground Truth', 'Predicted Image']
        
        for i in range(3):
            plt.subplot(1, 3, i+1)
            plt.title(title[i])
            # Getting the pixel values in the [0, 1] range to plot.
            plt.imshow(tf.keras.utils.array_to_img(display_list[i]* 0.5 + 0.5),cmap='gray')
            #plt.imshow(display_list[i] * 0.5 + 0.5)
            plt.axis('off')
        plt.show()
        
        ax6 = fig6.add_subplot(111)
        ax6.clear()
        ax6.imshow(tf.keras.utils.array_to_img(display_list[0]* 0.5 + 0.5),cmap='gray')
        ax6.axis('off')
        plt6_windows.draw()

        ax7 = fig7.add_subplot(111)
        ax7.clear()
        ax7.imshow(tf.keras.utils.array_to_img(display_list[1]* 0.5 + 0.5),cmap='gray')
        ax7.axis('off')
        plt7_windows.draw()

        ax8 = fig8.add_subplot(111)
        ax8.clear()
        ax8.imshow(tf.keras.utils.array_to_img(display_list[2]* 0.5 + 0.5),cmap='gray')
        ax8.axis('off')
        plt8_windows.draw()
      
    def compile(self, d_optimizer, g_optimizer, s_loss_fn, k_loss_fn, metrics_s, metrics_d):
        super(GAN_GGO, self).compile()
        self.g_optimizer = g_optimizer
        self.s_loss_fn = s_loss_fn
        self.TP_S = metrics_s[0]
        self.FN_S = metrics_s[1]
        self.FP_S = metrics_s[2]
        
        self.d_optimizer = d_optimizer
        self.k_loss_fn = k_loss_fn
        #self.TP_D = metrics_d[0]
        #self.FN_D = metrics_d[1]
        #self.FP_D = metrics_d[2]
  
    @property
    def metrics(self):
        # We list our `Metric` objects here so that `reset_states()` can be
        # called automatically at the start of each epoch
        # or at the start of `evaluate()`.
        # If you don't implement this property, you have to call
        # `reset_states()` yourself at the time of your choosing.
        return [self.metrics1]

    def discriminator_loss(self, disc_real_output, disc_generated_output):
        real_loss = self.k_loss_fn(tf.ones_like(disc_real_output), disc_real_output)
        generated_loss = self.k_loss_fn(tf.zeros_like(disc_generated_output), disc_generated_output)
        total_disc_loss = real_loss + generated_loss

        return total_disc_loss

    def generator_loss(self, disc_generated_output, gen_output, target):
        
        ####ini mengikuti tutorial pix2pix
        gan_loss = self.s_loss_fn(gen_output, target)
        # Mean absolute error
        l1_loss = tf.reduce_mean(tf.abs(target - gen_output))
        total_gen_loss = gan_loss + (self.LAMBDA * l1_loss)

        return total_gen_loss, gan_loss, l1_loss

    @tf.function
    def train_step(self, input_image, target, step):

        # Train the discriminator and generator
        with tf.GradientTape() as gen_tape, tf.GradientTape() as disc_tape:
            # generate prediction from generator
            pred_gen_output = self.generator(input_image, training=True)

            # generate prediction from discriminator
            pred_disc_real_output = self.discriminator([input_image, target], training=True)
            pred_disc_generated_output = self.discriminator([input_image, pred_gen_output], training=True)

            disc_loss = self.discriminator_loss(pred_disc_real_output, pred_disc_generated_output)
            gen_total_loss, gen_gan_loss, gen_l1_loss = self.generator_loss(pred_disc_generated_output, pred_gen_output, target)

            # Compute gradients
            generator_gradients = gen_tape.gradient(gen_total_loss,
                                                  generator.trainable_variables)
            discriminator_gradients = disc_tape.gradient(disc_loss,
                                                       discriminator.trainable_variables)
            # Update weights
            self.g_optimizer.apply_gradients(zip(generator_gradients,
                                                  generator.trainable_variables))
            self.d_optimizer.apply_gradients(zip(discriminator_gradients,
                                                      discriminator.trainable_variables))

        # Compute our own metrics
        self.TP_S.update_state(target, pred_gen_output)
        self.FN_S.update_state(target, pred_gen_output)
        self.FP_S.update_state(target, pred_gen_output)
        #self.TP_D.update_state(pred_disc_real_output, pred_disc_generated_output)
        #self.FN_D.update_state(pred_disc_real_output, pred_disc_generated_output)
        #self.FP_D.update_state(pred_disc_real_output, pred_disc_generated_output)

        return disc_loss, gen_gan_loss, pred_gen_output

    def fit(self, train_ds, test_ds,checkpoint, epochs):
        global checkpoint_prefix 
        example_input, example_target = next(iter(test_ds.take(1)))
        start = time.time()
        # Keep results for plotting
        train_g_loss_results = []
        train_d_loss_results = []
        train_g_metrics_TP_results = []
        train_g_metrics_FN_results = []
        train_g_metrics_FP_results = []
        DSC_score_metrics = []

        train_d_metrics_TP_results = []
        train_d_metrics_FN_results = []
        train_d_metrics_FP_results = []
        dsc_idx=0
        for epoch in range(epochs):
            epoch_g_loss_avg = tf.keras.metrics.Mean()
            epoch_d_loss_avg = tf.keras.metrics.Mean()
            st.insert(END, "================================================================\n")
            st.insert(END, "Start of epoch %d\n" % (epoch,))
            
            for step, (input_image, target) in train_ds.repeat().take(self.batch_size).enumerate():
                start = time.time()

                self.generate_images(generator, example_input, example_target)
                d_loss, g_loss, img_segm = self.train_step(input_image, target, step)

                #update loss setiap jaringan 
                epoch_g_loss_avg.update_state(g_loss)
                epoch_d_loss_avg.update_state(d_loss)

                if step % 10 == 0:
                    checkpoint.save(file_prefix = checkpoint_prefix)
                    # Print metrics
                    st.insert(END, "discriminator loss at step %d: %.2f" % (step, d_loss))
                    st.insert(END, "adversarial loss at step %d: %.2f" % (step, g_loss)) 
                 
            # Display metrics at the end of each epoch.
            train_g_loss_results.append(epoch_g_loss_avg.result())
            #train_g_metrics_TP_results.append(self.TP_S.result())
            #train_g_metrics_FN_results.append(self.FN_S.result())
            #train_g_metrics_FP_results.append(self.FP_S.result())

            #calculate DSC
            dsc_up = 2*self.TP_S.result()
            dsc_dw = dsc_up + self.FN_S.result() + self.FP_S.result()
            dsc_score = dsc_up/dsc_dw
            DSC_score_metrics.append(dsc_score)
            
            train_d_loss_results.append(epoch_d_loss_avg.result())
            #train_d_metrics_TP_results.append(self.TP_D.result())
            #train_d_metrics_FN_results.append(self.FN_D.result())
            #train_d_metrics_FP_results.append(self.FP_D.result())
            
            st.insert(END, "Training s loss over epoch: {}\n" .format(epoch_g_loss_avg.result()))
            st.insert(END, "Nilai TP Jaringan S: {}\n" .format(self.TP_S.result()))
            st.insert(END, "Nilai FN Jaringan S: {}\n" .format(self.FN_S.result()))
            st.insert(END, "Nilai FP Jaringan S: {}\n" .format(self.FP_S.result()))
            st.insert(END, "Nilai DSC Jaringan S: {}\n" .format(DSC_score_metrics[dsc_idx]))
            st.insert(END, "Training d loss over epoch: {}\n" .format(epoch_d_loss_avg.result()))
            #st.insert(END, "Nilai TP Jaringan C: {}\n" .format(self.TP_D.result()))
            #st.insert(END, "Nilai FN Jaringan C: {}\n" .format(self.FN_D.result()))
            #st.insert(END, "Nilai FP Jaringan C: {}\n" .format(self.FP_D.result()))
            st.insert(END, 'Time for epoch {} is {} sec\n'.format(epoch + 1, time.time()-start))
            st.insert(END, "================================================================\n")

            # Reset training metrics at the end of each epoch
            self.TP_S.reset_states()
            self.FN_S.reset_states()
            self.FP_S.reset_states()

            #self.TP_D.reset_states()
            #self.FN_D.reset_states()
            #self.FP_D.reset_states()
            
            epoch_g_loss_avg.reset_states()
            epoch_d_loss_avg.reset_states()
            dsc_idx +=1

        return train_g_loss_results, train_d_loss_results, DSC_score_metrics
```

```python
def pipeline_datas():
    global dataset, dataset_target, imgs_train,imgs_test, train_batches, test_batches, imgs_mask_train, orig_imgs_mask_test
    global trainIdx, testIdx

    indices = np.arange(0,len(dataset),1)

    imgs_train, imgs_test, \
    imgs_mask_train, orig_imgs_mask_test,\
    trainIdx, testIdx = train_test_split(dataset, dataset_target, indices,test_size=0.10)

    #augment datas
    augmented,aug_label,augmented_2,aug_label_2 = [],[], [],[]
    for i in range(len(imgs_train)):
        augmented += [Augment().call(imgs_train[i], imgs_mask_train[i])[0]]
        aug_label += [Augment().call(imgs_train[i], imgs_mask_train[i])[1]]
        augmented_2 += [Augment().call(imgs_train[i], imgs_mask_train[i])[2]]
        aug_label_2 += [Augment().call(imgs_train[i], imgs_mask_train[i])[3]]

    print(len(imgs_train))
    imgs_train = imgs_train+augmented+augmented_2
    imgs_mask_train = imgs_mask_train+aug_label+aug_label_2
    
    fill = "Data Latih Augmentasi :\t {}\n Data Uji Augmentasi :\t {}".format(len(imgs_train), len(imgs_mask_train))
    showinfo(title='Jumlah Data masukan', message=fill)
    print(len(imgs_train))

    train_datasets = tf.data.Dataset.from_tensor_slices((imgs_train,imgs_mask_train))
    test_datasets = tf.data.Dataset.from_tensor_slices((imgs_test, orig_imgs_mask_test))

    BUFFER_SIZE = 1000
    BATCH_SIZE = 16

    train_batches = train_datasets.cache().shuffle(BUFFER_SIZE).batch(BATCH_SIZE).repeat().prefetch(buffer_size=tf.data.AUTOTUNE)
    test_batches = test_datasets.batch(BATCH_SIZE)
    
```
